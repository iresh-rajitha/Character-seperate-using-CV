```python
# prompt: write a simple get request this request contain html page and I need to extract h2 tag content

import requests
from bs4 import BeautifulSoup

# for loop 1 to 195
# for page in range(1, 196):
#   url = f"https://sinhala-fonts.org/fonts/{page}"  # Replace with the actual URL

base_url = "https://sinhala-fonts.org"
font_names = []
page = 1
for page in range(1, 196):
  if page == 1:
    page = ""
  else:
    page = f"{page}"
  url = f"https://sinhala-fonts.org/fonts/{page}"  # Replace with the actual URL
  response = requests.get(url)
  soup = BeautifulSoup(response.content, "html.parser")
  h2_tags = soup.find_all("h2")
  for h2 in h2_tags:
    # get a tag detail inside this h2 tag
    a = h2.find('a')
    # print(a.get('href'))
    font_names.append(a.get('href').split('/')[-1])

print(font_names)
```

```python
# prompt: create download folder if not exist
import os

# Create the download folder if it doesn't exist
download_folder = "downloads"
if not os.path.exists(download_folder):
  os.makedirs(download_folder)
```

```python
import threading
import time

failed_fonts_names = []

def download_content(font):
  url = f"https://sinhala-fonts.org/fonts/{font}"
  filename = f"downloads/{font}.zip"
  try:
    response = requests.get(url)
    response.raise_for_status()

    with open(filename, 'wb') as file:
      file.write(response.content)
    print(f"Downloaded content to {filename}")

  except requests.exceptions.RequestException as e:
    failed_fonts_names.append(font)
    print(f"Error downloading content: {e}")

  downloaded_fonts.append(filename.split('/')[-1].replace('.zip', ''))

downloaded_fonts = []

# # Limit the number of threads
# max_threads = 10  # Adjust this value as needed
# semaphore = threading.Semaphore(max_threads)

threads = []
for font in font_names:
  thread = threading.Thread(target=download_content, args=(font,))
  threads.append(thread)
  thread.start()
  time.sleep(0.1)  # Add a small delay to avoid overloading the server

while any(thread.is_alive() for thread in threads):
  completed = len(downloaded_fonts)
  total = len(font_names)
  print(f"Progress: {completed}/{total} ({completed/total*100:.2f}%)")
  time.sleep(1)

for thread in threads:
  thread.join()
```

```python
# prompt: get size of downloads folder in MB (mega bytes)

import os

def get_folder_size_mb(folder_path):
  """Calculates the size of a folder in MB."""
  total_size = 0
  for dirpath, dirnames, filenames in os.walk(folder_path):
    for f in filenames:
      fp = os.path.join(dirpath, f)
      total_size += os.path.getsize(fp)
  return total_size / (1024 * 1024)  # Convert bytes to MB

download_folder = "downloads"
folder_size_mb = get_folder_size_mb(download_folder)
print(f"Size of '{download_folder}' folder: {folder_size_mb:.2f} MB")
```
